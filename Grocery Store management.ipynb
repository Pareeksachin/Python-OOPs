```python
class GroceryStore:
    def __init__(self):
        self.inventory = {}
        self.cart = {}
    
    def add_item(self,item,quantity,price):
        if item in self.inventory:
            self.inventory[item]["quantity"]+=quantity
        else:
            self.inventory[item] = {"quantity":quantity,"price":price}
    def remove_item(self,item,quantity):
        if item in self.inventory:
            if quantity >= self.inventory[item]["quantity"]:
                del self.inventory[item]
            else:
                self.inventory[item]["quantity"]-=quantity
    def add_to_cart(self,item,quantity):
        if item in self.inventory and quantity <= self.inventory[item]["quantity"]:
            if item in self.cart:
                self.cart[item]["quantity"]+=quantity
            else:
                self.cart[item]={"quantity":quantity,"price":self.inventory[item]["price"]}
            self.inventory[item]["quantity"]-=quantity
            print(f"{quantity} {item}(s) added to the cart.")
        else:
            print("Item is not avilable or quantity exceeds the avilable stock")
    def remove_item_cart(self,item,quantity):
        if item in self.cart:
            if quantity >= self.cart[item]["quantity"]:
                del self.cart[item]
            else:
                self.cart[item]["quantity"]-=quantity
                self.inventory[item]["quantity"]+= quantity
            print(f"{quantity} {item}(s) removed from the cart")
        else:
            print("item is not present in the cart")
            
    def calculateTotal():
        total = 0 
        if item in self.cart:
            price = self.cart[item]["price"]
            quantity = self.cart[item]["quantity"]
            total = total + price*quantity
        print(total)
            
    
    
    
    
    def display_inventory(self):
        print("Current Inventory")
        print("-----------------")
        for item in self.inventory:
            quantity = self.inventory[item]["quantity"]
            price = self.inventory[item]["price"]
            print(f"{item} : {quantity} - ${price} per ")
            
    def display_cart(self):
        print("Your Cart")
        print("-----------")
        for item in self.cart:
            quantity = self.cart[item]["quantity"]
            price = self.cart[item]["price"]
            print(f"{item} : {quantity} - ${price} per")
        

def grocery_management():
    store = GroceryStore()
    while True:
        print("============= Grocery Store Management ==========")
        print("""1 . Add items to the Inventory \n
                2. Remove items from Inventory
                3. Add item to the cart
                4. Remove item from cart
                5. Calculate Total
                6. Display Inventory
                7. Display cart
                8 . Exit """)
        choice = input("Enter your choice: ")
        if choice == "1":
            item = input("Enter item Name: ")
            quantity = int(input("Enter Quantity : "))
            price = float(input("Enter Price : "))
            store.add_item(item,quantity,price)
        elif choice =="2":
            item = input("Enter item Name : ")
            quantity = int(input("Enter Quantity :"))
            store.remove_item(item,quantity)
            
        elif choice =="3":
            item =  input("Enter item Name : ")
            quantity = int(input("Enter Quantity : "))
            store.add_to_cart(item,quantity)
        elif choice =="4":
            item = input("Enter item Name :")
            quantity = int(input("Enter Quantity : "))
            store.remove_from_cart(item,quantity)
        elif choice =="5":
            store.calculateTotal()
        elif choice == "6":
            store.display_inventory()
        elif choice == "7":
            store.display_cart()
        elif choice=="8":
            print("Thanks for visiting our Grocery Store, Have a good Day")
            break
        else:
            print("You have choose a Invalid Option ! please choose Correct option")
            
grocery_management()          
```
